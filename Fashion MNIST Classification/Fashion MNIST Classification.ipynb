## **First Deep Learning Project End To End | Fashion-MNIST Classification**

# **Project name: Fashion MNIST Classification**

```python
import keras
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.fashion_mnist.load_data()
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

```python
X_train[0]
```

```python
y_train[0]
```

```python
class_labels = ["T-shirt/top","Trouser","Pullover","Dress","Coat","Sandal","Shirt","Sneaker","Bag","Ankle boot"]
'''
0 => T-shirt/top 
1 => Trouser 
2 => Pullover 
3 => Dress 
4 => Coat 
5 => Sandal 
6 => Shirt 
7 => Sneaker 
8 => Bag 
9 => Ankle boot '''
```

```python
plt.imshow(X_train[0], cmap ="Greys")
```

```python
plt.figure(figsize=(16,16))
for i in range(25):
  plt.subplot(5,5,i+1)
  plt.imshow(X_train[i],cmap="Greys")
  plt.axis('off')
  plt.title(class_labels[y_train[i]]+"="+str(y_train[i]), fontsize=20)
```

# Feature Scalling

```python
X_train = X_train/255
X_test = X_test/255
```

```python
X_train[0]
```

# Build Neural Network

```python
model = keras.models.Sequential([
                         keras.layers.Flatten(input_shape=[28,28]),
                         keras.layers.Dense(units=32, activation='relu'),
                         keras.layers.Dense(units=10, activation='softmax')
])
```

```python
model.summary()
```

```python
model_2 = keras.models.Sequential([
                         keras.layers.Flatten(input_shape=[28,28]),
                         keras.layers.Dense(units=300, activation='relu'),
                         keras.layers.Dense(units=200, activation='relu'),
                         keras.layers.Dense(units=100, activation='relu'),
                         keras.layers.Dense(units=10, activation='softmax')
])
```

```python
model_2.summary()
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=5)
```

```python
model.fit(X_train, y_train, epochs=25)
```

```python
model_2.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=25)
```

```python
model_2.fit(X_train, y_train, epochs=25)
```

# Test and Evaluate Neural Network Model

```python
model.evaluate(X_test,y_test)
```

```python
model_2.evaluate(X_test,y_test)
```

```python
y_pred = model.predict(X_test)
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, [ np.argmax(i) for i in y_pred])
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(cm, annot=True, fmt = "d")
```

# Classification Report

```python
from sklearn.metrics import classification_report
cr = classification_report(y_test, [ np.argmax(i) for i in y_pred], target_names = class_labels,)
```
